# Alphabet Soup Venture Optimization

# Overview
The nonprofit foundation Alphabet Soup wants a tool that can help it select the applicants for funding with the best chance of success in their ventures.

# Results
Neural Network Model:
- Accuracy: 79.85%
- Loss: 58.77%

  ## Data Preprocessing
    - Previously successful ventures are the target variable for this model.
    - Name, application type, affiliation, classification, use case, organization, income amount, and ask amount are all features for this model.
    - The EIN variable should be removed from this model as it is neither a target or feature.

  ## Compiling, Training, and Evaluating the Model
    - The neural network model included 5 hidden layers, 447 input dimensions, and two activation functions.  With many input features a higher number of neurons was selected for the first hidden layer, 150. More layers were added to slowly reduce the number of neurons (150, 100, 50, 25, 12) until the output layer was reached. ReLu activation was chosen for all hiden layers for it's ability to learn complex patterns in data and focus on the most relevant features with sigmoid being used for the output layer. 100 epochs were run to give the model time to learn and self-correct.
    - The goal for this model was 75% accuracy. While model performance was met, there is room for improvement.
    - To increase model performance, the 'EIN' variable was removed as it did not add value to the dataset. The The 'NAME', 'APPLICATION TYPE', and 'CLASSIFICATION' features were scaled down by placing lower counts of each into an 'other' category. Each non-numeric feature was then encoded using 'get_dummies'. The number of hidden layers and neurons within the layers were adjusted to increase the accuracy of the model.

# Summary
The neural network model predicts whether applicants will be successful if funded by Alphabet Soup with 79.85% accuracy. While this indicates the model performs reasonably well, there is room for improvement, and exploring different models may help to improve results. Gradient Boosting Machines (GBMs) are effective for classification problems with structured data or Random Forest which is known fore it's bability to handle a large number of features.
